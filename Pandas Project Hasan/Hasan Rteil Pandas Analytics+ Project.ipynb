```python
import os
import glob
import pandas as pd
os.chdir("C:\\Users\\user\\Pandas Project Hasan") #Current working directory that contains your dataset and code file
os.getcwd()
```

#### To-Do - Use glob to match the pattern "csv"

```python
extension = 'csv'
all_filenames = [i for i in glob.glob(f"*{extension}")]
print(all_filenames)
```

#### To-Do - Combine all the files in the generated list above and export to a CSV
Tips - Use the `pd.concat` method

```python
combine_files = pd.concat(
    map(pd.read_csv, (all_filenames)), ignore_index = True)
print(combine_files)

#Export to csv

combine_files.to_csv('combinefiles.csv', index=False)
```
